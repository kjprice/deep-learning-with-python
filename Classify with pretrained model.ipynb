```python
import keras
from keras.preprocessing import image

from keras.applications.resnet50 import ResNet50
from keras.applications import resnet50

from keras.applications.xception import Xception
from keras.applications import xception

from keras.applications.vgg16 import VGG16
from keras.applications import vgg16

from keras.applications.vgg19 import VGG19
from keras.applications import vgg19

from keras.applications.inception_v3 import InceptionV3
from keras.applications import inception_v3

import numpy as np
```

```python
modelXception = Xception(weights='imagenet')
modelResNet50 = ResNet50(weights='imagenet')
modelVGG16 = VGG16(weights='imagenet')
modelVGG19 = VGG19(weights='imagenet')
modelInceptionV3 = InceptionV3(weights='imagenet')
```

```python
def prepare_image(img, application):
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = application.preprocess_input(x)
    return x
```

```python
import matplotlib.pyplot as plt
def predict_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_large = image.load_img(img_path, target_size=(299, 299))
    plt.imshow(img)

    x1 = prepare_image(img, xception)
    x2 = prepare_image(img, resnet50)
    x3 = prepare_image(img, vgg16)
    x4 = prepare_image(img, vgg19)
    x5 = prepare_image(img_large, inception_v3)

    preds1 = modelXception.predict(x1)
    preds2 = modelResNet50.predict(x2)
    preds3 = modelVGG16.predict(x3)
    preds4 = modelVGG19.predict(x4)
    preds5 = modelInceptionV3.predict(x5)
    print('Predicted xception:', xception.decode_predictions(preds1, top=3)[0])
    print('Predicted resnet50:', resnet50.decode_predictions(preds2, top=3)[0])
    print('Predicted vgg16:', vgg16.decode_predictions(preds3, top=3)[0])
    print('Predicted vgg19:', vgg19.decode_predictions(preds4, top=3)[0])
    print('Predicted inception_v3:', inception_v3.decode_predictions(preds5, top=3)[0])
```

```python
tiger_image = 'data/tiger.jpg'
predict_image(tiger_image)
```

```python
kids_in_car = 'data/kids_in_car.jpg'
predict_image(kids_in_car)
```

```python
family = 'data/family_portrait.jpg'
predict_image(family)
```

```python
architect = 'data/architect.jpg'
predict_image(architect)
```
